# 🚀 Step 8: Content Added to Netflix Over Time

### Platform Acquisition Pacing and Library Growth Velocity

</div>

---

## 📖 Explanation

### Content Added to Netflix Over Time

The `date_added` column shows when a movie or TV show was added to Netflix. Analyzing this information helps us understand how Netflix's content library has grown over time.

This analysis answers the research question:
> **"How has Netflix expanded its content library over the years?"**

By examining the number of titles added each year, we can identify periods of rapid growth and content acquisition.

---

<div style="background-color:#f5f5f5; padding:15px; border-left:6px solid #e50914;">

### 💡 Ingestion Timeline Workflow

Unlike tracking the original production year, isolating the onboarding dates requires extracting the year component specifically from the day content went live on the platform. This timeline highlights the operational scaling of Netflix's distribution network, showing when licensing machinery and ingestion pipelines went into overdrive to build a massive digital catalog.

</div>

```python
# Extract year from date_added

df['year_added'] = df['date_added'].dt.year

# Count titles added per year
content_added = df['year_added'].value_counts().sort_index()

print(content_added)
```

```python
# Content Added to Netflix Over Time

plt.figure(figsize=(14,6))

sns.lineplot(
    x=content_added.index,
    y=content_added.values,
    marker='o',
    linewidth=3
)

plt.title('Netflix Content Added Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')

plt.grid(True, alpha=0.3)

plt.show()
```

## 📖 Explanation

### Correlation Analysis and Heatmap

Correlation analysis is used to determine the strength and direction of relationships between numerical variables. A correlation coefficient ranges from $-1$ to $+1$:

* **$+1$** $\rightarrow$ Perfect positive relationship
* **$0$** $\rightarrow$ No relationship
* **$-1$** $\rightarrow$ Perfect negative relationship

In this dataset, we will examine the relationship between:
* **Release Year**
* **Year Added to Netflix**

This analysis helps answer the research question:
> **"Does newer content tend to be added to Netflix more quickly or frequently than older, archival content?"**

---

<div style="background-color:#f5f5f5; padding:15px; border-left:6px solid #e50914;">

### 💡 Statistical Covariance Workflow

To perform this evaluation, we first extract the numeric year components from both fields. Running a Pearson correlation matrix on these features reveals how closely a movie's production timeline aligns with its digital platform debut, and plotting the matrix as a visual heatmap makes the structural connection immediately clear.

</div>

```python
# Select numerical columns

numeric_df = df[['release_year', 'year_added']]

# Correlation matrix

correlation_matrix = numeric_df.corr()

print(correlation_matrix)
```

```python
# Heatmap Visualization
# Correlation Heatmap

plt.figure(figsize=(8,5))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    linewidths=1,
    fmt='.2f'
)

plt.title('Correlation Heatmap')

plt.show()
```
