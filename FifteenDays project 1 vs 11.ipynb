## 🎨 Feature Engineering Vectorization

By mapping the temporal attributes into standard data metrics, we expand the raw logs to extract high-value time dimensions across multiple analysis axes:

```python
# ==========================================
# Create a copy of the dataset before
# performing any cleaning operations.
# ==========================================

clean_df = df.copy()


# ==========================================
# Check the shape of the copied dataset.
# ==========================================

print("Dataset Shape Before Cleaning")
print(clean_df.shape, '\n')


# ==========================================
# Check for duplicate rows.
# ==========================================

print("Duplicate Rows Before Cleaning")
print(clean_df.duplicated().sum(), '\n')


# ==========================================
# Remove duplicate rows.
# ==========================================

clean_df = clean_df.drop_duplicates()


# ==========================================
# Display the shape after removing duplicates.
# ==========================================

print("Dataset Shape After Removing Duplicates")
print(clean_df.shape)
```

```python
# ==========================================
# Check for missing values again.
# ==========================================

print("Missing Values")
print(clean_df.isnull().sum(), '\n')


# ==========================================
# Calculate the total number of missing values.
# ==========================================

print("Total Missing Values")
print(clean_df.isnull().sum().sum(), '\n')


# ==========================================
# Display the percentage of missing values.
# ==========================================

missing_percentage = (clean_df.isnull().sum() / len(clean_df)) * 100

print("Percentage of Missing Values")
print(missing_percentage)
```

```python
# ==========================================
# Convert the InvoiceDate column into
# DateTime format.
# ==========================================

clean_df['InvoiceDate'] = pd.to_datetime(clean_df['InvoiceDate'])


# ==========================================
# Confirm that the conversion was successful.
# ==========================================

print("Updated Data Types")
print(clean_df.dtypes, '\n')


# ==========================================
# Display the first five rows.
# ==========================================

print(clean_df.head())
```

```python
# ==========================================
# Create new columns from InvoiceDate.
# ==========================================

clean_df['Year'] = clean_df['InvoiceDate'].dt.year
clean_df['Month'] = clean_df['InvoiceDate'].dt.month
clean_df['MonthName'] = clean_df['InvoiceDate'].dt.month_name()
clean_df['Day'] = clean_df['InvoiceDate'].dt.day
clean_df['DayOfWeek'] = clean_df['InvoiceDate'].dt.day_name()


# ==========================================
# Display the first five rows to verify
# the newly created columns.
# ==========================================

print(clean_df.head())
```

```python
# ==========================================
# Display the updated dataset information
# after cleaning.
# ==========================================

print("Cleaned Dataset Information")
print(clean_df.info(), '\n')


# ==========================================
# Display the final shape of the dataset.
# ==========================================

print("Final Dataset Shape")
print(clean_df.shape)
```

## 🎨 Exploratory Scope & Core Inquiries

Our descriptive queries and mathematical summaries are specifically configured to pinpoint five key metrics:

```python
# ==========================================
# Calculate the total revenue generated.
# ==========================================

print("Total Revenue Generated")
print("₦{:,.2f}".format(clean_df['TotalPrice'].sum()), '\n')


# ==========================================
# Calculate the average revenue per transaction.
# ==========================================

print("Average Revenue Per Transaction")
print("₦{:,.2f}".format(clean_df['TotalPrice'].mean()), '\n')


# ==========================================
# Calculate the highest transaction value.
# ==========================================

print("Highest Transaction Value")
print("₦{:,.2f}".format(clean_df['TotalPrice'].max()), '\n')


# ==========================================
# Calculate the lowest transaction value.
# ==========================================

print("Lowest Transaction Value")
print("₦{:,.2f}".format(clean_df['TotalPrice'].min()))
```

```python
# ==========================================
# Display the Top 10 Products
# with the highest quantity sold.
# ==========================================

top_products = clean_df.groupby('Description')['Quantity'].sum().sort_values(ascending=False)

print("Top 10 Products by Quantity Sold")
print(top_products.head(10), '\n')


# ==========================================
# Display the Top 10 Products
# generating the highest revenue.
# ==========================================

top_revenue_products = clean_df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False)

print("Top 10 Products by Revenue")
print(top_revenue_products.head(10))
```

```python
# ==========================================
# Display the Top 10 Countries
# by Total Revenue.
# ==========================================

country_sales = clean_df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)

print("Top Countries by Revenue")
print(country_sales, '\n')


# ==========================================
# Display the number of transactions
# made by each country.
# ==========================================

country_transactions = clean_df['Country'].value_counts()

print("Transactions by Country")
print(country_transactions)
```

```python
# ==========================================
# Display the Top 10 Customers
# based on total spending.
# ==========================================

top_customers = clean_df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)

print("Top 10 Customers by Spending")
print(top_customers.head(10), '\n')


# ==========================================
# Display customers with the
# highest number of transactions.
# ==========================================

customer_transactions = clean_df['CustomerID'].value_counts()

print("Top Customers by Number of Transactions")
print(customer_transactions.head(10))
```

```python
# ==========================================
# Display monthly revenue.
# ==========================================

monthly_sales = clean_df.groupby('MonthName')['TotalPrice'].sum()

print("Monthly Revenue")
print(monthly_sales, '\n')


# ==========================================
# Display yearly revenue.
# ==========================================

yearly_sales = clean_df.groupby('Year')['TotalPrice'].sum()

print("Yearly Revenue")
print(yearly_sales)
```

```python
# ==========================================
# Bar Chart of the Top 10 Products
# by Quantity Sold.
# ==========================================

top_products.head(10).plot(
    kind='bar',
    figsize=(10,6),
    title='Top 10 Products by Quantity Sold'
)

plt.xlabel("Products")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# ==========================================
# Bar Chart of the Top 10 Products
# by Revenue.
# ==========================================

top_revenue_products.head(10).plot(
    kind='bar',
    figsize=(10,6),
    title='Top 10 Products by Revenue'
)

plt.xlabel("Products")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# Bar Chart showing Total Revenue
# generated by each Country.
# ==========================================

country_sales.plot(
    kind='bar',
    figsize=(10,6),
    title='Revenue by Country'
)

plt.xlabel("Country")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# ==========================================
# Pie Chart showing Revenue Share
# by Country.
# ==========================================

country_sales.plot(
    kind='pie',
    figsize=(8,8),
    autopct='%1.1f%%'
)

plt.ylabel("")
plt.title("Revenue Share by Country")
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# Line Chart showing Monthly Revenue.
# ==========================================

monthly_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = monthly_sales.reindex(monthly_order)

monthly_sales.plot(
    kind='line',
    marker='o',
    figsize=(10,6),
    title='Monthly Revenue Trend'
)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()


# ==========================================
# Bar Chart showing Yearly Revenue.
# ==========================================

yearly_sales.plot(
    kind='bar',
    figsize=(7,5),
    title='Yearly Revenue'
)

plt.xlabel("Year")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
# ==========================================
# Histogram showing the distribution
# of Total Transaction Amount.
# ==========================================

clean_df['TotalPrice'].plot(
    kind='hist',
    bins=30,
    figsize=(10,6),
    title='Distribution of Transaction Amount'
)

plt.xlabel("Transaction Amount")
plt.tight_layout()
plt.show()


# ==========================================
# Box Plot showing outliers in
# transaction amounts.
# ==========================================

clean_df.boxplot(column='TotalPrice', figsize=(6,6))

plt.title("Box Plot of Transaction Amount")
plt.show()
```
