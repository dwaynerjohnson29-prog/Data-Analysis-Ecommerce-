## 🎨 Feature Engineering Layout

Parsing the contract timestamp array expands our relational columns, unlocking high-value timeline analytics across multiple seasonal segments:

```python
# ==========================================
# Create a copy of the dataset before
# performing any cleaning operations.
# ==========================================

clean_df = df.copy()


# ==========================================
# Display the shape of the dataset
# before cleaning.
# ==========================================

print("Dataset Shape Before Cleaning")
print(clean_df.shape, '\n')


# ==========================================
# Check for duplicate rows.
# ==========================================

print("Duplicate Rows Before Cleaning")
print(clean_df.duplicated().sum(), '\n')


# ==========================================
# Remove duplicate rows.
# ==========================================

clean_df = clean_df.drop_duplicates()


# ==========================================
# Display the shape after
# removing duplicate rows.
# ==========================================

print("Dataset Shape After Removing Duplicates")
print(clean_df.shape)
```

```python
# ==========================================
# Check for missing values
# in each column.
# ==========================================

print("Missing Values")
print(clean_df.isnull().sum(), '\n')


# ==========================================
# Display the total number
# of missing values.
# ==========================================

print("Total Missing Values")
print(clean_df.isnull().sum().sum(), '\n')


# ==========================================
# Display the percentage
# of missing values.
# ==========================================

missing_percentage = (clean_df.isnull().sum() / len(clean_df)) * 100

print("Percentage of Missing Values")
print(missing_percentage)
```

```python
# ==========================================
# Convert the JoinDate column
# into DateTime format.
# ==========================================

clean_df['JoinDate'] = pd.to_datetime(clean_df['JoinDate'])


# ==========================================
# Display the updated
# data types.
# ==========================================

print("Updated Data Types")
print(clean_df.dtypes, '\n')


# ==========================================
# Display the first five rows
# of the cleaned dataset.
# ==========================================

print("First Five Rows")
print(clean_df.head())
```

```python
# ==========================================
# Create new columns from JoinDate.
# ==========================================

clean_df['YearJoined'] = clean_df['JoinDate'].dt.year
clean_df['MonthJoined'] = clean_df['JoinDate'].dt.month
clean_df['MonthName'] = clean_df['JoinDate'].dt.month_name()
clean_df['Day'] = clean_df['JoinDate'].dt.day
clean_df['DayOfWeek'] = clean_df['JoinDate'].dt.day_name()


# ==========================================
# Display the first five rows
# to verify the new columns.
# ==========================================

print("Updated Dataset")
print(clean_df.head())
```

```python
# ==========================================
# Display the cleaned dataset
# information.
# ==========================================

print("Cleaned Dataset Information")
print(clean_df.info(), '\n')


# ==========================================
# Display the final shape
# of the cleaned dataset.
# ==========================================

print("Final Dataset Shape")
print(clean_df.shape)
```

```python
# ==========================================
# Display the total number of players.
# ==========================================

print("Total Number of Players")
print(len(clean_df), '\n')


# ==========================================
# Display the average age of players.
# ==========================================

print("Average Player Age")
print(round(clean_df['Age'].mean(), 2), '\n')


# ==========================================
# Display the average overall rating.
# ==========================================

print("Average Overall Rating")
print(round(clean_df['Overall'].mean(), 2), '\n')


# ==========================================
# Display the average potential rating.
# ==========================================

print("Average Potential Rating")
print(round(clean_df['Potential'].mean(), 2), '\n')


# ==========================================
# Display the average market value.
# ==========================================

print("Average Market Value")
print("₦{:,.2f}".format(clean_df['MarketValue'].mean()))
```

```python
# ==========================================
# Display the Top 10 highest-rated players.
# ==========================================

top_players = clean_df.sort_values(by='Overall', ascending=False)

print("Top 10 Highest Rated Players")
print(top_players[['PlayerName', 'Club', 'Overall']].head(10), '\n')


# ==========================================
# Display the Top 10 most valuable players.
# ==========================================

valuable_players = clean_df.sort_values(by='MarketValue', ascending=False)

print("Top 10 Most Valuable Players")
print(valuable_players[['PlayerName', 'Club', 'MarketValue']].head(10))
```

```python
# ==========================================
# Display the clubs with the highest
# average overall rating.
# ==========================================

club_rating = clean_df.groupby('Club')['Overall'].mean().sort_values(ascending=False)

print("Average Overall Rating by Club")
print(club_rating, '\n')


# ==========================================
# Display the Top 10 clubs with the
# highest total market value.
# ==========================================

club_value = clean_df.groupby('Club')['MarketValue'].sum().sort_values(ascending=False)

print("Top Clubs by Total Market Value")
print(club_value.head(10))
```

```python
# ==========================================
# Display the number of players
# from each nationality.
# ==========================================

nationality = clean_df['Nationality'].value_counts()

print("Players by Nationality")
print(nationality, '\n')


# ==========================================
# Display the number of players
# by playing position.
# ==========================================

positions = clean_df['Position'].value_counts()

print("Players by Position")
print(positions)
```

```python
# ==========================================
# Display the Top 10 goal scorers.
# ==========================================

top_goal_scorers = clean_df.sort_values(by='Goals', ascending=False)

print("Top 10 Goal Scorers")
print(top_goal_scorers[['PlayerName', 'Club', 'Goals']].head(10), '\n')


# ==========================================
# Create a new column called
# Goal Contributions.
# ==========================================

clean_df['GoalContributions'] = clean_df['Goals'] + clean_df['Assists']


# ==========================================
# Display the Top 10 players with
# the highest goal contributions.
# ==========================================

goal_contributions = clean_df.sort_values(by='GoalContributions', ascending=False)

print("Top Players by Goal Contributions")
print(goal_contributions[['PlayerName', 'Club', 'GoalContributions']].head(10))
```
