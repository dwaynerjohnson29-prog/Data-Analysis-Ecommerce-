# 🔥 Correlation Analysis

---

Correlation analysis helps identify relationships between numerical variables in the dataset.

A correlation coefficient ranges from -1 to +1:

- A value close to +1 indicates a strong positive relationship.
- A value close to -1 indicates a strong negative relationship.
- A value close to 0 indicates little or no relationship.

This analysis aims to determine which factors have the strongest influence on property prices.

---

````python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Matrix')

plt.show()
````

# 🧪 Hypothesis Testing: Does Property Type Affect Property Price?

---

Hypothesis testing is a statistical method used to determine whether there is enough evidence to support a claim about a population.

In this analysis, we investigate whether property type significantly influences property prices.

---

## Null Hypothesis (H₀)

There is no significant difference in property prices among different property types.

In other words, property type does not affect property price.

---

## Alternative Hypothesis (H₁)

There is a significant difference in property prices among different property types.

In other words, property type affects property price.

---

## Significance Level

The significance level (α) used for this test is:

α = 0.05

---

### Decision Rule

- If the p-value < 0.05, Reject H₀.
- If the p-value ≥ 0.05, Fail to Reject H₀.

A One-Way ANOVA test is used because we are comparing the mean prices of more than two property types.

---

````python
from scipy.stats import f_oneway

# Separate prices by property type

apartment = df[df['Property_Type'] == 'Apartment']['Price']
villa = df[df['Property_Type'] == 'Villa']['Price']
duplex = df[df['Property_Type'] == 'Duplex']['Price']
bungalow = df[df['Property_Type'] == 'Bungalow']['Price']

# ANOVA Test

f_statistic, p_value = f_oneway(
    apartment,
    villa,
    duplex,
    bungalow
)

print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
````

## Hypothesis Test Result

---

The ANOVA test produced a p-value less than 0.05.

Since the p-value is less than the significance level (0.05), the null hypothesis (H₀) is rejected.

---

### Decision

Reject H₀

---

### Conclusion

There is sufficient statistical evidence to conclude that property type significantly affects property price.

Therefore, different property types tend to have different average market prices.
