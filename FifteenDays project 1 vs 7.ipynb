## 📊 Core Visualization Implementations (Seaborn & Matplotlib)

We utilize the power of `seaborn` and `matplotlib` to convert our loan database arrays into clear, analytical visual assets:

### 1. Count Plot (`sns.countplot()`)
The count plot maps discrete categories into straightforward vertical or horizontal bars, explicitly showing the record volume inside each classification group. It provides an immediate visual benchmark for:
* 💼 **Application Volume:** Comparing employment status counts.
* 🏡 **Property Profiles:** Evaluating homeownership distribution records.

### 2. The `hue` Parameter
The `hue` parameter is a critical layer in multi-variable analysis. It tells Seaborn to break down each parent bar into distinct, color-coded sub-segments based on a secondary categorical field.
> 💡 **Application:** Setting `hue="Loan_Status"` instantly splits your demographic bar graphs into grouped blocks showing exactly how many loans were *Fully Paid*, *Current*, or *Charged Off* side-by-side.

### 3. Histogram (`sns.histplot()`)
The histogram splits continuous numerical data into uniform intervals called **bins**, counting how many data observations drop into each range. We use this plot to map features like **Annual Income** to figure out:
* 💰 If our core demographic falls mostly into lower or higher salary ranges.
* 📈 Whether the incoming volume curve is balanced, normal, or skewed toward a specific limit.

### 4. Box Plot (`sns.boxplot()`)
The box plot provides a standardized five-number summary of continuous credit metrics (like **Loan Amount** or **Credit Score**). It displays your data distribution through its core quartiles:
* 🟩 **The Median:** The central mark tracking the exact midpoint of your loan values.
* 📏 **The Interquartile Range (IQR):** The main body container wrapping the middle 50% of values.
* 🛡️ **Outliers:** Distinct points plotted outside the whiskers, highlighting unusually massive or extreme loan applications.
