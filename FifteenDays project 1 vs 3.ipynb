# 📊 Step 5: Top 10 Countries with Highest COVID-19 Cases

### Aggregating Global Infection Spreads and Volumetric Rankings

</div>

---

## 📖 Explanation

### Top 10 Countries with the Highest COVID-19 Cases

One of the major objectives of this study is to identify the countries most affected by COVID-19.

This analysis aggregates the total number of confirmed cases for each country and ranks them from highest to lowest.

The visualization provides a clear comparison of the countries with the largest burden of COVID-19 infections.

---

<div style="background-color:#f5f5f5; padding:15px; border-left:6px solid #2a5298;">

### 💡 Ranking and Segmentation Workflow

Isolating high-density subsets through group aggregation and value ranking allows you to strip away background noise, highlighting the critical distribution hubs that account for the vast majority of cumulative global data patterns.

</div>

```python
# Top 10 countries by total cases

top_cases = df.groupby("Country")["Total_Cases"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_cases.values, y=top_cases.index)

plt.title("Top 10 Countries with Highest COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")

plt.show()
```

# 📈 COVID-19 Cases Trend Over Time

### Mapping Longitudinal Infection Curves and Public Health Impact Eras

</div>

---

## 📖 Explanation

### Trend of COVID-19 Cases Over Time

Analyzing the trend of COVID-19 cases over time helps understand how the pandemic evolved during different periods. By observing changes in the number of new cases, it is possible to identify periods of rapid transmission, peaks in infections, and potential declines resulting from public health interventions.

This visualization provides a clear picture of the progression of COVID-19 cases over time and highlights important patterns in the spread of the virus.

---

<div style="background-color:#f5f5f5; padding:15px; border-left:6px solid #2a5298;">

### 💡 Time-Series Dynamics Workflow

Converting chronological data coordinates into continuous lines exposes velocity shifts and seasonal patterns. Tracking these longitudinal metrics lets you map infection milestones against containment implementations to measure intervention efficacy.

</div>

```python
# Total new cases by date

cases_trend = df.groupby("Date")["New_Cases"].sum().reset_index()

plt.figure(figsize=(14,6))
plt.plot(cases_trend["Date"], cases_trend["New_Cases"])

plt.title("COVID-19 New Cases Trend Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)

plt.show()
```

# 🎛️ Correlation Analysis

### Mapping Multi-Variable Matrices and Statistical Directional Strengths

</div>

---

## 📖 Explanation

### Correlation Analysis

Correlation analysis is used to measure the strength and direction of relationships between numerical variables in the dataset. It helps identify whether increases or decreases in one variable are associated with changes in another variable.

A correlation heatmap provides a visual representation of these relationships, making it easier to identify strong positive or negative correlations among COVID-19 indicators such as cases, deaths, vaccinations, and population.

---

<div style="background-color:#f5f5f5; padding:15px; border-left:6px solid #2a5298;">

### 💡 Heatmap Matrix Workflow



Computing Pearson coefficient matrices standardizes linear dependencies onto a clear $[-1, 1]$ scale. Visualizing this distribution via a color-coded matrix allows you to instantly extract structural insights, isolate multi-collinearity issues, and isolate feature pairings optimized for predictive modeling.

</div>

```python
# Select numerical columns

numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix

correlation_matrix = numeric_df.corr()

# Plot heatmap

plt.figure(figsize=(10,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("COVID-19 Correlation Heatmap")

plt.show()
```
