## 🔬 Measuring Linear Dependencies

Correlation analysis is used to measure the strength and direction of the linear relationship between numerical variables in a dataset. Understanding these statistical alignments helps data scientists identify which variables are positively or negatively associated with one another, signaling predictive patterns that can support strategic business decision-making.

In this section, we expose how hidden operational metrics impact each other—such as determining whether long booking lead times are linked to high cancellation occurrences, or how the length of a guest's stay shifts the final Average Daily Rate (ADR).

---

## 🧮 Understanding the Correlation Coefficient ($r$)

The relationships are calculated on a scale from **-1.0 to +1.0**, which can be visualized through a structured data frame matrix:



<div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(280px, 1fr)); gap: 16px; margin-top: 20px; margin-bottom: 25px;">
    <div style="background: #0f172a; border: 1px solid #1e293b; border-left: 4px solid #10b981; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        📈 <strong>Positive Correlation ($r > 0$):</strong> As one variable increases, the other moves up simultaneously (e.g., total stay duration increasing as total weeknight counts scale up).
    </div>
    <div style="background: #0f172a; border: 1px solid #1e293b; border-left: 4px solid #ef4444; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        📉 <strong>Negative Correlation ($r < 0$):</strong> As one variable increases, the paired parameter drops cleanly in the opposite direction.
    </div>
    <div style="background: #0f172a; border: 1px solid #1e293b; border-left: 4px solid #64748b; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        ⚪ <strong>Neutral Bounds ($r \approx 0$):</strong> No noticeable or linear relationship exists between the targeting data columns.
    </div>
</div>

---

> 🗺️ **Matrix Roadmap:** The Python block below extracts numeric categories automatically using `.select_dtypes()`, builds a Pearson correlation matrix with `.corr()`, and renders an annotated multi-color visual heatmap via `sns.heatmap()`.

```python
# Select only numerical columns
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Correlation Matrix
correlation = numeric_df.corr()

# Display Correlation Matrix
correlation

# Plot Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
