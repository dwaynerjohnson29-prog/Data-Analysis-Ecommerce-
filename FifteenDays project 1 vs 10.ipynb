## <i class='bx bx-git-branch'></i> Mapping Variable Interdependencies

Correlation analysis is a statistical technique used to measure the strength and direction of the linear relationship between numerical variables in a dataset. It helps identify whether specific workforce variables increase or decrease together, assisting data analysts in discovering hidden behavioral patterns within corporate structures. 

A correlation heatmap provides a visual representation of these matrix relationships, making it significantly easier to identify strong positive or negative correlations. Understanding these relationships enables organizations to make precise, data-driven decisions regarding employee performance, satisfaction, compensation tiers, and long-term retention frameworks.

---

## <i class='bx bx-grid-alt'></i> Linear Variance Checkpoints

The relationships are mathematically calculated between the boundaries of **-1.0 and +1.0**:



<div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(280px, 1fr)); gap: 16px; margin-top: 20px; margin-bottom: 25px;">
    <div style="background: #141d26; border: 1px solid #2e4057; border-left: 4px solid #f4d35e; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        <i class='bx bx-trending-up' style='color: #f4d35e; font-size: 1.2rem; vertical-align: middle; margin-right: 4px;'></i> <strong>Positive Alignment ($r > 0$):</strong> Variables move in the exact same direction (e.g., higher monthly income correlating positively with total years at the company).
    </div>
    <div style="background: #141d26; border: 1px solid #2e4057; border-left: 4px solid #f4d35e; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        <i class='bx bx-trending-down' style='color: #f4d35e; font-size: 1.2rem; vertical-align: middle; margin-right: 4px;'></i> <strong>Inverse Alignment ($r < 0$):</strong> Variables move in opposite directions (e.g., higher job satisfaction correlates with a downward trend in attrition metrics).
    </div>
    <div style="background: #141d26; border: 1px solid #2e4057; border-left: 4px solid #f4d35e; padding: 16px; border-radius: 6px; color: #e2e8f0; font-size: 0.92rem; font-family: sans-serif;">
        <i class='bx bx-radio-circle' style='color: #f4d35e; font-size: 1.2rem; vertical-align: middle; margin-right: 4px;'></i> <strong>Neutral Bounds ($r \approx 0$):</strong> The targeted attributes display no linear co-dependency or discernible mathematical impact on one another.
    </div>
</div>

> <i class='bx bx-slider-alt'></i> **Matrix Roadmap:** The cell below isolates the data frame's numeric variables using `.select_dtypes(include=['int64', 'float64'])`, generates the Pearson correlation matrix via `.corr()`, and renders an annotated Seaborn heatmap using `sns.heatmap()`.

```python
# ==========================================================
# 📊 SELECTING NUMERICAL COLUMNS
# ==========================================================

print("="*70)
print("📊 SELECTING NUMERICAL COLUMNS")
print("="*70)

numeric_df = df.select_dtypes(include=["int64", "float64"])

display(numeric_df.head())

# ==========================================================
# 🔥 GENERATING THE CORRELATION MATRIX
# ==========================================================

print("="*70)
print("🔥 GENERATING THE CORRELATION MATRIX")
print("="*70)

correlation = numeric_df.corr()

display(correlation)

# ==========================================================
# 📈 VISUALIZING THE CORRELATION HEATMAP
# ==========================================================

print("="*70)
print("📈 VISUALIZING THE CORRELATION HEATMAP")
print("="*70)

plt.figure(figsize=(12,8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Employee Attrition Dataset")
plt.show()

# ==========================================================
# 📦 DISPLAYING THE STRONGEST CORRELATIONS
# ==========================================================

print("="*70)
print("📦 DISPLAYING THE STRONGEST CORRELATIONS")
print("="*70)

display(correlation.unstack().sort_values(ascending=False).drop_duplicates())
```
